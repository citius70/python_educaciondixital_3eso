## Ejercicio resuelto

**¿Puede entrar al concierto?**

Programa que pide la edad del usuario y comprueba si puede entrar solo al concierto.

```python
edad = int(input("¿Cuántos años tienes? "))

if edad >= 18:
    print("Puedes entrar solo al concierto ✅")
else:
    print("Necesitas ir acompañado 👨‍👩‍👦")
```
