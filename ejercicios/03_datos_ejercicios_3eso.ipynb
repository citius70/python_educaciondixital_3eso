## Ejercicio 6: Evita errores al mezclar tipos ⚠️

**Objetivo**: Aprender a convertir texto a número para evitar errores comunes.

**Problema común**:
```python

```

**Solución correcta**:
```python

edad_numero = int(edad)  # Convertimos texto a número
print(edad_numero + 5)   # ✅ Funciona: 20
```
