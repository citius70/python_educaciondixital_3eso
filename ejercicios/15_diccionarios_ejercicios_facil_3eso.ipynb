## Ejercicio 1: Ficha de un videojuego

Se quiere guardar la información de un videojuego en un diccionario.

```python
videojuego = {
    "titulo": "Aventura Espacial",
    "genero": "Plataformas",
    "precio": 19.99,
    "jugadores": 2
}
```

**Pasos**
1. Copia el diccionario `videojuego`.
2. Imprime el diccionario completo.
3. Imprime, usando un mensaje con `print()`, el **título** del juego.
4. Imprime el **precio** del juego.
5. Imprime el número de **jugadores**.
6. Cambia el precio del juego a `14.99` (porque está de oferta) usando `update()`.
7. Vuelve a imprimir el diccionario completo para comprobar que el precio ha cambiado.

**Pista**: para acceder a un valor se usa `diccionario["clave"]`.

## Ejercicio 2: Recorrer un diccionario de asignaturas

Se tiene un diccionario con las **asignaturas** de un alumno y la **nota** obtenida en cada una.

```python
notas = {
    "Matemáticas": 7,
    "Tecnología": 9,
    "Inglés": 6,
    "Educación Física": 8
}
```

**Pasos**
1. Copia el diccionario `notas`.
2. Usando un bucle `for`, recorre el diccionario con `.items()` y muestra una línea por cada asignatura, con el formato:

   ```
   La nota de Matemáticas es 7
   La nota de Tecnología es 9
   La nota de Inglés es 6
   La nota de Educación Física es 8
   ```

3. Después del bucle, calcula cuántas asignaturas tiene el diccionario usando `len()` y muéstralo con `print()`.

**Pista**: `for asignatura, nota in notas.items():`
