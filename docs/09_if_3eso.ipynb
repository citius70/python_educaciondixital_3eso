# 📗 Condicionales: if, elif y else

## Aprende a tomar decisiones en Python

Las **estructuras condicionales** permiten que tu programa tome decisiones y ejecute código diferente según la situación. Son como las decisiones que tomas cada día:

- "**Si** llueve, cojo el paraguas"
- "**Si** tengo hambre, como algo"
- "**Si** es fin de semana, duermo más"
    - "**Si no** es fin de semana, hago los deberes"

En Python usamos `if`, `elif` y `else` para crear estas decisiones. ¡Vamos a verlo! 🚀

## La estructura `if` (si...)

La estructura `if` ejecuta un bloque de código **solo si una condición es verdadera**.

**Sintaxis:**
```python
if condicion:
    # código que se ejecuta si la condición es verdadera
```

:::{warning} ⚠️ Importante: La indentación
El código que va después del `if` **debe ir con sangría** (4 espacios o una tabulación). Esto indica que pertenece al `if`.
:::

### Ejemplo 1: Número positivo

````python
numero = 5

if numero > 0:
    print("El número es positivo ✅")

print("El programa continúa aquí")
# Esta última línea SIEMPRE se ejecuta, esté o no dentro del if
````

:::{tip} 💡 ¿Qué pasa aquí?
1. Python comprueba si `numero > 0` (¿5 es mayor que 0?)
2. Como es **verdadero**, ejecuta el código con sangría: `print("El número es positivo ✅")`
3. Después continúa con el resto del programa

**Prueba:** Cambia `numero = 5` por `numero = -3` y vuelve a ejecutar. ¿Qué pasa?
:::

### Ejemplo 2: Acceso a un concierto

````python
edad = 16

if edad >= 18:
    print("Puedes entrar al concierto 🎵")

# Si edad es menor de 18, no se imprime nada
````

## La estructura `if...else` (si... si no...)

El `else` ("si no") ejecuta un bloque de código **cuando la condición del `if` es falsa**.

**Sintaxis:**
```python
if condicion:
    # código si la condición es verdadera
else:
    # código si la condición es falsa
```

### Ejemplo 3: Aprobado o suspendido

````python
nota = 4.5

if nota >= 5:
    print("¡Aprobado! 🎉")
else:
    print("Suspendido 😞 Ánimo para la próxima")

# Prueba cambiando la nota a 7
````

### Ejemplo 4: Par o impar

````python
numero = 7

if numero % 2 == 0:  # El operador % da el resto de la división
    print("El número es par")
else:
    print("El número es impar")
````

:::{note} 📚 Operador módulo `%`
El operador `%` devuelve el **resto** de una división:
- `7 % 2` → 1 (porque 7 dividido entre 2 es 3 y sobra 1)
- `8 % 2` → 0 (porque 8 dividido entre 2 es 4 exacto, no sobra nada)

Por eso, si `numero % 2 == 0`, el número es par (no sobra nada al dividir entre 2).
:::

## La estructura `if...elif...else` (múltiples opciones)

El `elif` ("else if", que significa "si no, si...") permite evaluar **varias condiciones** en orden.

**Sintaxis:**
```python
if condicion_1:
    # código si condicion_1 es verdadera
elif condicion_2:
    # código si condicion_1 es falsa y condicion_2 es verdadera
elif condicion_3:
    # código si condicion_1 y condicion_2 son falsas y condicion_3 es verdadera
else:
    # código si todas las condiciones anteriores son falsas
```

### Ejemplo 5: Clasificación de notas

````python
nota = 8.5

if nota >= 9:
    print("Sobresaliente 🌟")
elif nota >= 7:
    print("Notable 👍")
elif nota >= 5:
    print("Aprobado ✅")
else:
    print("Suspendido 📚")

# Con nota = 8.5, el resultado será "Notable 👍"
````

:::{tip} 💡 ¿Cómo funciona?
Python evalúa las condiciones **en orden**, de arriba hacia abajo:
1. ¿Es `nota >= 9`? No (8.5 no es mayor o igual a 9)
2. ¿Es `nota >= 7`? **Sí** (8.5 es mayor o igual a 7)
3. Ejecuta ese bloque y **se detiene** (no evalúa el resto)
:::

### Ejemplo 6: Clasificar un número

````python
numero = -3

if numero > 0:
    print("El número es positivo")
elif numero < 0:
    print("El número es negativo")
else:
    print("El número es cero")
````

## 📊 Operadores de comparación

Para crear condiciones, necesitas **comparar valores**. Estos son los operadores principales:

| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|----------|
| == | Igual a | 5 == 5 | `True` |
| != | Diferente de | 5 != 3 | `True` |
| > | Mayor que | 5 > 3 | `True` |
| < | Menor que | 5 < 3 | `False` |
| >= | Mayor o igual que | 5 >= 5 | `True` |
| <= | Menor o igual que | 5 <= 3 | `False` |

:::{warning} ⚠️ Cuidado: `=` vs `==`
- = → Asigna un valor a una variable (edad = 15)
- == → Compara si dos valores son iguales (edad == 15)

❌ if edad = 15: → ERROR  
✅ if edad == 15: → CORRECTO
:::

### Ejemplo 7: Comparar valores

````python
a = 5
b = 10

if a == b:
    print("a es igual a b")
else:
    print("a no es igual a b")  # Esto se imprimirá
````

````python
# Operador diferente !=
precio = 25

if precio != 20:
    print("El precio no es 20 euros")  # Esto se imprimirá
````

## 🔗 Combinar condiciones: `and` y `or`

A veces necesitas que se cumplan **varias condiciones a la vez**, o que se cumpla **al menos una**.

### Operador `and` (y)

Se deben cumplir **todas** las condiciones para que sea `True`.

**Ejemplo del mundo real:** "Puedo salir **si** he terminado los deberes **Y** he recogido mi cuarto"

````python
edad = 16
tiene_permiso = True

if edad >= 16 and tiene_permiso:
    print("Puedes ir a la excursión ✅")
else:
    print("No puedes ir ❌")
````

### Operador `or` (o)

Basta con que se cumpla **al menos una** condición para que sea `True`.

**Ejemplo del mundo real:** "Llevo chaqueta **si** hace frío **O** si llueve"

````python
es_fin_de_semana = True
es_festivo = False

if es_fin_de_semana or es_festivo:
    print("¡No hay clase! 🎉")
else:
    print("Toca ir al instituto")
````

### Ejemplo 8: Llevar paraguas

````python
estoy_afuera = True
esta_lloviendo = True

if estoy_afuera and esta_lloviendo:
    print("Lleva el paraguas ☂️")
else:
    print("No necesitas paraguas")
````

### Ejemplo 9: Combinar `and` y `or`

````python
# Descuentos en un cine
edad = 14
es_estudiante = True

if edad < 18 or es_estudiante:
    print("Tienes descuento del 30% 🎬")
else:
    print("Precio normal")
````

## 📝 Resumen

:::{important} Lo que has aprendido

✅ **Estructura `if`:**
- Ejecuta código solo si una condición es verdadera

✅ **Estructura `if...else`:**
- Ejecuta un código si la condición es verdadera, otro si es falsa

✅ **Estructura `if...elif...else`:**
- Evalúa múltiples condiciones en orden
- Se ejecuta solo el primer bloque cuya condición sea verdadera

✅ **Operadores de comparación:**
- `==`, `!=`, `>`, `<`, `>=`, `<=`

✅ **Combinar condiciones:**
- `and` → Todas las condiciones deben cumplirse
- `or` → Al menos una condición debe cumplirse

✅ **Regla importante:**
- El código dentro del `if` debe llevar **sangría** (indentación)
:::
