# 📗 Listas

Una **lista** es una colección **ordenada y modificable** de elementos (o valores).
Los elementos de una lista van encerrados entre corchetes `[ ]`.

## Crear y mostrar una lista

```python
frutas = ["manzana", "plátano", "cereza"]
print(frutas)
```

En este ejemplo, la lista `frutas` tiene 3 elementos.
El orden es importante: primero `"manzana"`, luego `"plátano"` y después `"cereza"`.

```{note} Características de las listas
- **Mutables**: se pueden cambiar después de crearlas (añadir, quitar, modificar elementos).
- **Ordenadas**: los elementos tienen una posición (índice) que empieza en 0.
- Pueden contener **cualquier tipo de dato**: números, textos o una mezcla.
```

## Índices: acceder a los elementos

Podemos acceder a los elementos de una lista usando su **índice numérico**.
El primer elemento tiene índice `0`, el segundo `1`, etc.

```python
colores = ["rojo", "verde", "azul", "amarillo"]
primer_elemento = colores[0]   # "rojo"
segundo_elemento = colores[1]   # "verde"
ultimo_elemento = colores[3]    # "amarillo"
print(primer_elemento, segundo_elemento, ultimo_elemento)
```

En este ejemplo:
- `colores[0]` es `"rojo"`.
- `colores[1]` es `"verde"`.
- `colores[3]` es `"amarillo"`.

## Cambiar elementos de una lista

Como las listas son **mutables**, podemos cambiar sus valores:

```python
numeros =[64, 56, 62, 89]
numeros[2] = 99   # cambiamos el 62 por 99
print(numeros)    #[64, 56, 99, 89]
```

También podemos añadir nuevos elementos al final de la lista con `append()`:

```python
asignaturas = ["Matemáticas", "Lengua"]
asignaturas.append("Inglés")
print(asignaturas)  # ["Matemáticas", "Lengua", "Inglés"]
```

## Métodos básicos de listas

Partimos de esta lista:

```python
animales = ["perro", "gato", "elefante"]
```

Algunos métodos útiles:

1. **append(elemento)** → añade un elemento al final.

```python
animales.append("jirafa")
print(animales)  # ["perro", "gato", "elefante", "jirafa"]
```

2. **remove(elemento)** → borra la primera vez que aparece ese elemento.

```python
animales.remove("gato")
print(animales)  # ["perro", "elefante", "jirafa"]
```

3. **len(lista)** → devuelve cuántos elementos tiene la lista.

```python
cantidad = len(animales)
print(cantidad)  # 3
```

### Otros métodos útiles
Partimos de una lista con precios:

```python
precios = [10, 20, 15, 5, 10, 23]
```

1. **sort()** → ordena la lista de menor a mayor.
```python
precios.sort()
print(precios)  # [5, 10, 10, 15, 20, 23]
```
2. **reverse()** → invierte el orden de la lista.
```python
precios.reverse()
print(precios)  # [23, 10, 5, 15, 20, 10]
```
3. **index(elemento)** → devuelve el índice de la primera aparición del elemento.
```python
indice = precios.index(15)
print(indice)  # 2
```
4. **count(elemento)** → cuenta cuántas veces aparece el elemento en la lista.
```python
cantidad = precios.count(10)
print(cantidad)  # 2
```

## Recorrer una lista con `for`

Muchas veces queremos hacer algo con **cada elemento** de la lista.

```python
asignaturas = ["Matemáticas", "Lengua", "Inglés"]

for elemento in asignaturas:
    print("Tengo examen de", elemento)
```

Salida:

```text
Tengo examen de Matemáticas
Tengo examen de Lengua
Tengo examen de Inglés
```

## Slices (trozos de lista)

Un **slice** permite obtener una parte de la lista.

```python
numeros =[5, 4, 6, 8, 3]

primeros_dos = numeros[0:2]       # desde el inicio hasta el índice 1
primeros_tres = numeros[:3]       # desde el inicio hasta el índice 2 
algunos_numeros = numeros[1:4]    # desde el índice 1 hasta el 3 (no incluye el 4)

print(primeros_dos)               # Salida: [5, 4]
print(primeros_tres)              # Salida: [5, 4, 6]
print(algunos_numeros)            # Salida: [4, 6, 8]
```

## Resumen

- Una **lista** se escribe con corchetes: `[...]`.
- Las listas son **ordenadas** y **modificables**.
- El primer elemento tiene índice `0`.
- Métodos básicos importantes:
  - `append()` para añadir.
  - `remove()` para quitar.
  - `len()` para contar elementos.
- Podemos recorrer una lista con un bucle `for`.
