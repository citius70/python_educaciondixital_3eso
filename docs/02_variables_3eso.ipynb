# 📗 Variables en Python

En este tema aprenderás qué son las **variables** en programación, cómo crearlas y las reglas básicas para nombrarlas. ¡Verás que es muy sencillo! 😊

## ¿Qué es una variable?

```{admonition} Una variable es...
:class: note
... como una **caja** 📦 donde guardamos información en el ordenador. Esa información puede ser un número, un texto, o cualquier otra cosa que necesitemos.
```

Imagina que tienes una caja y le pones una **etiqueta** con un nombre (por ejemplo, *mis_puntos*). Dentro de esa caja guardas un número (por ejemplo, *100*). Cada vez que quieras saber cuántos puntos tienes, solo necesitas mirar dentro de la caja llamada *mis_puntos*.

**Crear una variable en Python es muy fácil:**

`nombre_de_la_variable = valor`

El símbolo `=` significa "guardar" o "asignar". Le estamos diciendo a Python: "Guarda este valor en esta variable".

Por ejemplo, si escribimos `mi_numero = 4`, estamos creando una variable llamada `mi_numero` y guardando el número `4` dentro. Es como poner una etiqueta "mi_numero" en una caja que contiene el número 4.

````python
# Creamos una variable
mi_numero = 4
````

````python
puntuacion = 86
puntuacion = 88  # Ahora la variable tiene otro valor
puntuacion = 90  # Vuelve a cambiar
````

:::{seealso} ¿Qué es print()? 👈
:class: dropdown
**print()** es un comando que le dice a Python: "¡Muestra esto en la pantalla!" Es muy útil para ver qué hay guardado en una variable.

En Python podemos guardar diferentes tipos de información:

````python
edad = 15
precio = 9.99
````

**Texto** (siempre entre comillas):

## Reglas para nombrar variables

En Python no podemos poner **cualquier nombre** a nuestras variables. Hay algunas reglas importantes que debemos seguir:

### 1️⃣ Python distingue mayúsculas y minúsculas

Para Python, `nombre`, `Nombre` y `NOMBRE` son tres variables **diferentes**.

````python
# Estas son tres variables distintas
nombre = "Carlos"
Nombre = "María"
NOMBRE = "Luis"
````

### 2️⃣ Debe empezar por letra o guion bajo

✅ Puede empezar por letra (a-z, A-Z) o guion bajo (_)

❌ **NO** puede empezar por número

````python
# Ejemplos correctos
edad = 15
_secreto = 42

# Ejemplo INCORRECTO (da error)
# 1edad = 15  # ¡Esto no funciona!
````

### 3️⃣ Solo letras, números y guion bajo

✅ Puedes usar letras, números y guion bajo (_)

❌ **NO** puedes usar espacios, tildes ni símbolos raros (@, #, %, etc.)

````python
# Ejemplos correctos
mi_edad = 15
numero_2 = 100

# Ejemplos INCORRECTOS
# mi edad = 15      # ¡Espacios no!
# mi-edad = 15      # ¡El guion medio no!
# mi@variable = 15  # ¡Símbolos no!
````

### 4️⃣ No usar palabras reservadas de Python

Python tiene algunas palabras especiales que **NO** podemos usar como nombres de variables. Por ejemplo: `if`, `for`, `while`, `print`, etc.

````python
# Ejemplo INCORRECTO
# if = 10  # ¡No podemos usar 'if'!

# Ejemplo correcto
condicion = 10  # Usamos otro nombre
````

### 5️⃣ Usa nombres que tengan sentido

Es mejor usar nombres que expliquen qué contiene la variable. Así tu código es más fácil de entender.

````python
# 😊 Nombre claro y descriptivo
numero_de_estudiantes = 25

# 😕 Nombre poco claro
n = 25  # ¿Qué significa 'n'?
````

## Estilo recomendado: snake_case 🐍

En Python, la forma **recomendada** de escribir nombres de variables es usar **snake_case**: todo en minúsculas y separando palabras con guion bajo (_).

````python
# Ejemplos de snake_case
mi_nombre = "Pedro"
numero_favorito = 7
edad_del_usuario = 15
````

:::{seealso} Otros estilos que existen (avanzado)
:class: dropdown

Aunque **snake_case** es lo recomendado en Python, existen otros estilos:

- **camelCase**: `miNombreCompleto = "Ana"`
- **PascalCase**: `MiNombreCompleto = "Ana"`

De momento, usa siempre **snake_case** para tus variables.
:::

## Resumen: Reglas básicas

```{admonition} Reglas para nombrar variables 📝
:class: tip
- ✅ Deben empezar por **letra** o **guion bajo** (_)
- ❌ **NO** pueden empezar por número
- ❌ **NO** pueden tener espacios, tildes ni símbolos raros
- ❌ **NO** usar palabras reservadas de Python (`if`, `for`, etc.)
- 🐍 Usa **snake_case**: todo en minúsculas con guiones bajos
- 💡 Elige nombres que tengan sentido

| ✅ Variables válidas | ❌ Variables NO válidas |
|---------------------|------------------------|
| `edad` | `1edad` |
| `mi_nombre` | `mi nombre` |
| `numero_2` | `número-2` |
| `_secreto` | `for` |
```
