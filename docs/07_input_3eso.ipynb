# 📗 Función `input()`

En este notebook vas a aprender a usar la función `input()` para que tus programas puedan **pedir datos al usuario desde el teclado**.


## 1. ¿Qué hace `input()`?

::: {note} `input()`
:class: info

La función `input()` permite al usuario escribir datos con el teclado y devuelve ese valor como **texto** (cadena de caracteres). Normalmente lo guardamos en una variable.

```python
variable = input("mensaje")
```
:::

La función `input()` hace que el programa **se pare**, muestre un mensaje en pantalla y espere a que el usuario escriba algo y pulse Intro.


::: {danger} Importante
:class: danger

El texto que escribe el usuario se **devuelve siempre como texto** (tipo `str`).

- Aunque el usuario escriba un número, `input()` lo devuelve como texto, no como número.
- Si quieres usar ese número para hacer cálculos, tendrás que convertirlo a número usando funciones como `int()` o `float()`.
:::


## 2. Primer ejemplo: leer un nombre

En este ejemplo se pide al usuario su nombre y se muestra por pantalla:

```python
# Pedimos el nombre al usuario y lo mostramos
name = input("¿Cuál es tu nombre? ")
print("Hola,", name)
```


En este código:

- `input("¿Cuál es tu nombre? ")` muestra el mensaje y espera a que el usuario escriba su nombre.
- Lo que el usuario escribe se guarda en la variable `name`.
- Después se muestra un saludo usando `print()`.


## 3. La función `input()` siempre devuelve texto

Es muy importante recordar que `input()` **siempre** devuelve una **cadena de texto** (`str`), aunque el usuario escriba números.

Si quieres trabajar con números (sumar, restar, etc.), tendrás que **convertir** ese texto a número entero (`int`) o decimal (`float`).


### 3.1. Convertir la entrada a número

Si pedimos la edad a una persona, nos interesa tratarla como número para poder hacer operaciones con ella (por ejemplo, sumar 1):

```python
# Pedimos la edad y la convertimos a entero
age = int(input("¿Cuántos años tienes? "))
print("El próximo año tendrás", age + 1)
print("Tipo de dato de age:", type(age))
```


Aquí ocurre lo siguiente:

- `input("¿Cuántos años tienes? ")` devuelve un texto, por ejemplo `"15"`.
- `int("15")` convierte ese texto al número entero `15`.
- Ahora es posible sumar `1` y obtener la edad del próximo año.


## 5. Ejemplo: sumar dos números introducidos por el usuario

Primero veremos qué pasa si **no** convertimos la entrada a número, y después la versión correcta.


### 5.1. Versión incorrecta (concatena texto)

```python
num_a = input("Introduce el primer número: ")
num_b = input("Introduce el segundo número: ")
resultado = num_a + num_b
print(resultado)
```

Si el usuario introduce 2 y 3, el resultado será `23`, porque está **juntando textos**, no sumando números.


### 5.2. Versión correcta (suma números)

```python
# Pedimos dos números y los sumamos correctamente
num_a = int(input("Introduce el primer número: "))
num_b = int(input("Introduce el segundo número: "))

resultado = num_a + num_b
print("La suma es:", resultado)
```

Ahora, si el usuario introduce 2 y 3, el resultado será `5`, porque estamos trabajando con números enteros.
