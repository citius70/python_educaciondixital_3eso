# 📗 Bucle `while`

## Repetir mientras se cumpla una condición

:::{note} **Bucle `while`**

Se utiliza para repetir un bloque de código **mientras una condición sea verdadera**.
La condición se comprueba al inicio de cada vuelta del bucle.
Cuando deja de cumplirse, el programa **sale del bucle**.
:::

> Importante: un `while` permite **repetir** un bloque de código un **número indeterminado** de veces.
> No sabemos cuántas veces se ejecutará (1, 2, 3, ...), depende de la condición.

## Sintaxis básica del bucle `while`

```python
while condicion:
    # bloque de código a ejecutar mientras la condición sea verdadera

# código que se ejecuta después de salir del bucle
```

- `while`: palabra clave que indica el inicio del bucle.
- `condicion`: expresión que puede ser `True` o `False`.
- Mientras la condición sea `True`, el bloque con sangría se repite.
- Cuando la condición pasa a ser `False`, se termina el bucle.

## Ejemplo 1: Contador creciente

````python
contador = 1

while contador <= 5:
    print(contador, "paso")
    contador += 1  # Esto es equivalente a: contador = contador + 1

print("Cuando no se cumpla la condición (contador igual a 6), el programa sigue aquí")
````

:::dropdown Explicación del programa
- Empezamos con `contador = 1`.
- El bucle `while` se ejecuta **mientras** `contador <= 5`.
- En cada vuelta se imprime el valor de `contador` y luego se suma 1.
- Cuando `contador` llega a 6, la condición `contador <= 5` deja de cumplirse y salimos del bucle.
:::


## Ejemplo 2: Cuenta atrás 🚀

````python
contador_atras = 5  # Empezamos en 5

print("Comienza la cuenta atrás")

while contador_atras > 0:  # Mientras el número sea mayor que 0
    print(contador_atras)
    contador_atras -= 1  # Restamos 1 en cada vuelta

print("¡Despegue!")
````

:::dropdown Explicación del programa
- El bucle cuenta hacia atrás desde 5 hasta 1.
- En cada vuelta se muestra el valor actual y se resta 1.
- Cuando `contador_atras` llega a 0, la condición `contador_atras > 0` es falsa y el bucle termina.
- Después se imprime el mensaje "¡Despegue!".
:::


## `while True` y `break` (idea avanzada)

> :warning: ATENCIÓN: Leer solo en el caso de que quieras aprender algo más avanzado.

En algunos casos queremos que el programa **pida datos al menos una vez** y luego decida si sigue o no.

Python no tiene un bucle `do-while` como otros lenguajes, pero podemos conseguir algo parecido usando:
- `while True:` para crear un bucle que, en principio, no termina.
- `break` para **salir del bucle** cuando queramos.

### Ejemplo 3: Sumar números hasta que se introduce 0

Queremos escribir un programa que **sume números introducidos por el usuario** hasta que el usuario escriba `0`.

- El programa debe pedir números **al menos una vez**.
- Si el número es 0, dejamos de pedir más.
- Si no es 0, lo vamos sumando.

````python
suma = 0

while True:
    numero = int(input("Introduce un número (0 para terminar): "))
    if numero == 0:
        break  # Salimos del bucle si el número es 0
    suma = suma + numero

print(f"La suma total es: {suma}")
````

:::dropdown Explicación del programa
- Empezamos con `suma = 0`.
- `while True:` crea un bucle que, en principio, nunca acaba.
- En cada vuelta pedimos un número al usuario.
- Si el número es 0, usamos `break` para salir del bucle.
- Si el número es distinto de 0, lo añadimos a `suma`.
- Al final mostramos la suma total.
:::


## 📝 Resumen

- El bucle `while` repite un bloque **mientras se cumpla una condición**.
- Es útil cuando **no sabemos** cuántas veces se repetirá el bucle.
- Es importante actualizar las variables dentro del bucle para que, en algún momento, la condición deje de cumplirse.
- `while True` junto con `break` permite crear bucles que piden datos hasta que el usuario decide terminar.
