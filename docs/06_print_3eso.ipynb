# 📗Función `print()`

La función `print()` sirve para **mostrar mensajes en la pantalla**.
La usaremos constantemente para ver resultados y comprobar si el programa hace lo que queremos.


## 1. Uso básico de `print()`

### 1.1. Mostrar texto

Para mostrar un mensaje en la pantalla:

- Escribimos `print("texto para mostrar")`.
- Dentro de los paréntesis, ponemos el texto **entre comillas**.

```python
print("¡Hola, Mundo!")
# Salida: ¡Hola, Mundo!
```


### 1.2. Mostrar el valor de una variable

Si queremos ver qué guarda una variable, escribimos su nombre **sin comillas** dentro de `print(variable)`.

```python
x = 10
print(x)
# Salida: 10
```


## 2. Combinar texto y variables

Podemos mostrar texto y variables en la misma línea.
Basta con separarlos con comas, y Python pondrá un espacio entre cada elemento.

```python
distancia = 10
print("Hay una separación de", distancia, "metros.")
# Salida: Hay una separación de 10 metros.
```

> Nota: A cada cosa que ponemos dentro de `print()` (texto, variable, etc.) la llamamos **argumento**.


## 3. F-strings

Muchas veces queremos construir frases que incluyan varias variables.
Los **f-strings** son una forma muy cómoda de hacerlo.

- Se escribe una `f` antes de las comillas.
- Dentro del texto, las variables van entre llaves `{ }`.

```python
nombre = "Alberto"
edad = 35
print(f"Hola, me llamo {nombre} y tengo {edad} años.")
# Salida: Hola, me llamo Alberto y tengo 35 años.
```

Con f-strings también podemos poner operaciones dentro de las llaves.

```python
x = 5
y = 3
print(f"La suma de {x} y {y} es {x + y}")
# Salida: La suma de 5 y 3 es 8
```



## 4. Imprimir en varias líneas

Podemos escribir un texto que ocupe varias líneas usando `\n`, que significa "salto de línea".

```python
print("Línea 1 \nLínea 2 \nLínea 3")
```

Salida:

```text
Línea 1 
Línea 2 
Línea 3
```


## 5. Algunas opciones extra de `print()`

Estas opciones se pueden mostrar como curiosidad, sin exigir que las dominen.

- `sep` cambia el separador entre los elementos.
- `end` cambia lo que se imprime al final de la línea.

```python
# Cambiar el separador entre nombres
print("Alberto", "Sara", "Ana", sep="-")
# Salida: Alberto-Sara-Ana

# Cambiar el final de línea
print("Hola", end=" ")
print("Mundo")
# Salida: Hola Mundo
```

También podemos imprimir listas directamente para ver su contenido.

```python
frutas = ["manzana", "banana", "naranja"]
print("Stock de frutas:", frutas)
# Salida: Stock de frutas: ['manzana', 'banana', 'naranja']
```
