# 📗 Introducción a las estructuras de datos

## ¿Qué es una estructura de datos?

En Python, las **estructuras de datos** son formas de **guardar y organizar varios datos juntos** (no solo uno) para poder usarlos de manera cómoda.

Por ejemplo, podemos guardar:
- Las notas de una clase.
- Una lista de nombres de personas.
- Los precios de distintos productos.

En este curso vamos a trabajar sobre todo con:

- **Listas (`list`)**
- **Diccionarios (`dict`)**

También existen otras estructuras (tuplas y conjuntos), que veremos solo de forma muy básica al final.

## Listas [`list`]

Una **lista** es una colección de elementos **ordenados** (con un orden) y **modificables** (podemos cambiar sus valores).

- Se escriben entre corchetes `[]`.
- Sus elementos van separados por comas.
- Podemos guardar números, textos, o una mezcla.

**Ejemplos:**

```python
edades =[12, 15, 14, 13]
nombres = ["Ana", "Manuel", "María"]
de_todo = [1, "hola", 3.14, "manzanas"]
```

## Diccionarios {`dict`}

Un **diccionario** guarda datos en forma de **parejas**: clave → valor.

Es como un diccionario de verdad:
- La **clave** sería la palabra.
- El **valor** sería la definición.

En Python:
- Se escriben entre llaves `{}`.
- Cada par se escribe como `clave: valor`.
- Las claves suelen ser textos, y deben ser **únicas**.

**Ejemplos:**

```python
edades = {"Ana": 13, "Luis": 15, "María": 14}
precios = {"manzana": 0.5, "banana": 0.3, "naranja": 0.7}
```

## Otras estructuras (solo para conocer los nombres)

Además de listas y diccionarios, Python tiene otras estructuras de datos:

1. **Tuplas (`tuple`)**
   - Parecidas a las listas, pero **no se pueden modificar**.
   - Se escriben entre paréntesis `()`.
   - Ejemplos:
     ```python
     dias_semana = ("lunes", "martes", "miércoles", "jueves", "viernes")
     coordenadas = (40.7128, -74.0060)
     ```

2. **Conjuntos (`set`)**
   - Colecciones **sin orden** y **sin elementos repetidos**.
   - Se usan, por ejemplo, para eliminar duplicados.
   - Ejemplo:
     ```python
     frutas = {"manzana", "banana", "naranja", "manzana"}
     # El conjunto resultante solo tendrá una "manzana"
     ```

## Resumen rápido

| Estructura            | Sintaxis típica     | ¿Modificable? | ¿Tiene orden? | ¿Permite repetidos?          |
|-----------------------|--------------------|---------------|---------------|------------------------------|
| Lista (`list`)        | `[1, 2, 3]`        | Sí            | Sí            | Sí                           |
| Diccionario (`dict`)  | `{"Ana": 13}`      | Sí            | Sí            | Claves no, valores sí        |
| Tupla (`tuple`)       | `(1, 2, 3)`        | No            | Sí            | Sí                           |
| Conjunto (`set`)      | `{1, 2, 3}`        | Sí (añadir/quitar) | No        | No                           |
