# 📗 Introducción a las estructuras de control

## Haz que tus programas tomen decisiones

Hasta ahora, tus programas en Python eran como seguir una receta de cocina: hacías una cosa, luego otra, siempre en el mismo orden.

Pero los programas realmente útiles necesitan ser **más inteligentes**:
- ¿Qué pasa si quieres que el programa haga cosas diferentes según la situación?
- ¿Y si necesitas repetir una acción muchas veces sin escribir el mismo código?

¡Para eso sirven las **estructuras de control**! 🚀

## 📝 Ejecución secuencial (lo normal)

Por defecto, Python ejecuta las instrucciones **de arriba hacia abajo**, una tras otra, como si leyeras un libro línea por línea.

**Ejemplo:**

````python
print("Paso 1: Enciendo el ordenador")
print("Paso 2: Abro Python")
print("Paso 3: Empiezo a programar")
````

Python ejecuta cada `print()` en orden, sin saltarse ninguno. Esto está bien para tareas simples, pero ¿qué pasa si quieres que el programa se adapte a diferentes situaciones?

:::{tip} 💡 Analogía del mundo real
Piensa en tu rutina matinal:
1. Te levantas
2. Desayunas
3. Vas al instituto

Pero... **¿y si llueve?** Entonces coges el paraguas. **¿Y si es sábado?** No vas al instituto.

Tu cerebro toma **decisiones** según la situación. ¡Eso es lo que hacen las estructuras de control!
:::

## 🎯 ¿Qué son las estructuras de control?

:::{note} Definición
Las **estructuras de control** son herramientas que permiten que tus programas:
- **Tomen decisiones** ("Si hace frío, llevo chaqueta")
- **Repitan acciones** ("Escucho mi canción favorita 10 veces seguidas")

Gracias a ellas, el código se vuelve **flexible e inteligente** 🧠
:::

Hay **dos tipos principales** de estructuras de control:

1. **Condicionales** → Para tomar decisiones
2. **Bucles** → Para repetir acciones

::::{grid} 1 1 2 2

:::{card}
:header: 🔀 1. Estructuras condicionales

Permiten que el programa **tome decisiones** y ejecute código diferente según la situación.

**Palabras clave:**
- `if` → "si pasa esto, haz..."
- `else` → "si no, haz..."
- `elif` → "y si pasa esto otro, haz..."

**Ejemplo del mundo real:**

```python
edad = 14

if edad >= 18:
    print("Puedes entrar al concierto")
else:
    print("Necesitas ir acompañado")
```

**Resultado:** "Necesitas ir acompañado"

Como `edad` es 14 (menor que 18), Python ejecuta el código del `else`.
:::

:::{card}
:header: 🔁 2. Bucles

Permiten **repetir acciones** muchas veces sin escribir el mismo código una y otra vez.

**Tipos de bucles:**
- `for` → Repite un número determinado de veces
- `while` → Repite mientras se cumpla una condición

**Ejemplo con `for`:**

```python
for i in range(3):
    print("¡Me gusta Python!")
```

**Resultado:**
```
¡Me gusta Python!
¡Me gusta Python!
¡Me gusta Python!
```

En lugar de escribir `print()` tres veces, el bucle lo repite automáticamente.
:::

## 🧪 Ejemplos prácticos

Vamos a ver ejemplos que puedes ejecutar y modificar:

### Ejemplo 1: Condicional simple con `if` y `else`

````python
# Sistema de aprobado/suspendido
nota = 7.5

if nota >= 5:
    print("¡Aprobado! 🎉")
else:
    print("Suspendido 😞 Ánimo para la próxima")

# Prueba cambiando el valor de 'nota' a 4 y vuelve a ejecutar
````

### Ejemplo 2: Múltiples opciones con `elif`

````python
# Clasificación de notas completa
nota = 8.5

if nota >= 9:
    print("Sobresaliente 🌟")
elif nota >= 7:
    print("Notable 👍")
elif nota >= 5:
    print("Aprobado ✅")
else:
    print("Suspendido 📚")

# Con nota = 8.5, el resultado será "Notable 👍"
````

### Ejemplo 3: Bucle `for` para repetir acciones

````python
# Cuenta atrás para lanzar un cohete
print("Iniciando cuenta atrás...")

for numero in range(5, 0, -1):  # Cuenta de 5 a 1
    print(numero)

print("🚀 ¡Despegue!")
````

### Ejemplo 4: Bucle `while` que se repite hasta cumplir una condición

````python
# Ganar puntos hasta alcanzar 100
puntos = 0

while puntos < 100:
    puntos += 25  # Ganas 25 puntos por nivel
    print("Puntos actuales:", puntos)

print("¡Has alcanzado 100 puntos! 🏆")
````
