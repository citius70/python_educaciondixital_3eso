# 🔁 Bucles

## Repetir acciones en Python

Los **bucles** en Python se utilizan para **ejecutar un bloque de código varias veces** sin tener que escribirlo repetido línea a línea.

Son muy útiles, por ejemplo, para:
- Contar del 1 al 10.
- Mostrar un mensaje varias veces.
- Recorrer una lista de datos.

Python proporciona dos tipos principales de bucles:

- `for`: repite el código un número **determinado** de veces (1, 2, 3, ... repeticiones).
- `while`: repite el código un número **indeterminado** de veces, **mientras se cumpla una condición**.
