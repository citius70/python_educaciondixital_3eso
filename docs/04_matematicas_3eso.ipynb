# 📗 Operaciones matemáticas en Python

En este tema vamos a aprender a usar Python como una **calculadora mejorada**.
Trabajaremos con operaciones básicas (suma, resta, multiplicación, división), potencias y algunas funciones matemáticas útiles como `abs()` o `round()`.


## Operadores matemáticos básicos

En Python podemos hacer operaciones como en la calculadora, usando símbolos muy parecidos.

- Suma: `+`
- Resta: `-`
- Multiplicación: `*`
- División normal: `/` (puede dar decimales)
- División entera: `//` (se queda solo con la parte entera)
- Módulo: `%` (da el resto de una división)
- Potencia: `**` (elevar un número, por ejemplo `2 ** 3 = 8`)


### Ejemplo de la suma y distintas formas de presentación


````python
# Definimos dos números
a = 10
b = 5

# Forma 1: imprimir directamente la suma
print(a + b)

# Forma 2: guardar el resultado en una variable
resultado = a + b
print(resultado)

# Forma 3: mostrar un mensaje explicando el resultado
print("La suma de a y b es:", a + b)
print(f"La suma de {a} y {b} es: {a + b}")
````

### Ejemplos de operaciones básicas


````python
# Operaciones básicas en Python
a = 20
b = 3

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b}")        # puede dar decimales
print(f"División entera: {a} // {b} = {a // b}")  # solo la parte entera
print(f"Módulo (resto): {a} % {b} = {a % b}")
print(f"Potencia: {a} ** {b} = {a ** b}")
````

## Funciones matemáticas sencillas

Además de los operadores, Python tiene **funciones** que nos ayudan con cálculos.

Por ahora usaremos solo unas pocas muy útiles:

- `abs(x)`: valor absoluto de x (si es negativo lo pasa a positivo).
- `round(x, n)`: redondea x con n decimales.
- `max(lista)`: número más grande de una lista.
- `min(lista)`: número más pequeño de una lista.


````python
# Ejemplos de funciones sencillas

x = -6
print("Valor absoluto de x:", abs(x))

y = 4.567
print("Redondeo de y con 2 decimales:", round(y, 2))

numeros = [3, 10, 7, 2, 15]
print("Mayor número:", max(numeros))
print("Menor número:", min(numeros))
````

## Usar la biblioteca `math`

Cuando queremos operaciones un poco más avanzadas (por ejemplo usar π o sacar raíces cuadradas) podemos usar la biblioteca `math`.

Primero hay que importarla:

```python
import math
```

Algunas funciones fáciles de entender:

- `math.pi`: el número π (3.1415…).
- `math.sqrt(x)`: raíz cuadrada de x.
- `math.pow(x, y)`: x elevado a y (similar a `x ** y`).
- `math.ceil(x)`: redondea hacia arriba.
- `math.floor(x)`: redondea hacia abajo.


````python
import math

print("Valor de pi:", math.pi)

print("Raíz cuadrada de 16:", math.sqrt(16))
print("Potencia: 2 elevado a 5:", math.pow(2, 5))

x = 4.3
print("Redondeo hacia arriba de 4.3:", math.ceil(x))
print("Redondeo hacia abajo de 4.3:", math.floor(x))
print("Redondeo normal con round(4.3):", round(x))
````

## Ejemplo: área de un círculo

Recuerda que la fórmula del área de un círculo es:

Área = π · radio²

Vamos a calcularla con Python usando `math.pi`.


````python
import math

radio = 7  # en centímetros
area = math.pi * radio ** 2

# :.2f hace que se muestren solo 2 decimales
print(f"Área del círculo con radio {radio} cm: {area:.2f} cm2")
````
