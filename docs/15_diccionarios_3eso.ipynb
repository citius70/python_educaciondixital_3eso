# 📗 Diccionarios

Un **diccionario** es una forma de guardar información en parejas: una **clave** y un **valor**. 

Piensa en un diccionario de papel: buscas una palabra (la **clave**) y obtienes su significado (el **valor**). En Python funciona igual, pero las claves pueden ser palabras, números, etc.

* Las **claves** no se pueden repetir y no se pueden cambiar.
* Los **valores** sí se pueden cambiar cuando queramos.

## Creación de un diccionario
Los diccionarios se escriben entre llaves `{}`. Cada pareja clave-valor se separa con dos puntos `:`, y las parejas se separan entre sí con comas. La forma general es:

```python
mi_diccionario = {
    "clave1": valor1,
    "clave2": valor2,
    "clave3": valor3
}
```

Veamos un ejemplo concreto:


<img src=https://www.programiz.com/sites/tutorial2program/files/python_dictionary-example.png width="500">

````python
pais_capital = {'Alemania': 'Berlín',
                 'Canada': 'Otawa',
                 'Reino Unido': 'Londres'}

# Para acceder a un valor usamos: diccionario["clave"]
print(pais_capital['Reino Unido'])
````

Otro ejemplo, con valores de distinto tipo (números, texto, etc.):

```python  
empleada = {
    "nombre": "Alicia",
    "edad": 30,
    "ciudad": "Madrid"
}
```

## Métodos de manipulación de diccionarios
Los diccionarios son **mutables**: podemos añadir, modificar o eliminar valores. Vamos a ver los métodos más importantes con un ejemplo.

Empezamos creando un diccionario llamado `paisPoblacion`. Las **claves** serán nombres de **países** y los **valores** serán sus **poblaciones**.

````python
# Crear un diccionario de países y sus poblaciones
paisPoblacion = {
    'China': 1444216107,
    'India': 1393409038,
    'Estados Unidos': 331002651,
    'Indonesia': 273523615,
    'Pakistán': 225199937
}
````

1. **keys()**: nos da todas las **claves** del diccionario (en este caso, los países).

````python
claves = paisPoblacion.keys()
print("Claves del diccionario (países):", claves)
````

2. **values()**: nos da todos los **valores** del diccionario (en este caso, las poblaciones).

````python
valores = paisPoblacion.values()
print("Valores del diccionario (poblaciones):", valores) 
````

3. **items()**: nos da todas las **parejas clave-valor** del diccionario, juntas.

````python
elementos = paisPoblacion.items()
print("Elementos del diccionario (país, población):", elementos)  
````

4. **get(clave, valor_si_no_existe)**: nos da el valor de una clave. Si esa clave no existe, en lugar de dar un error, nos devuelve el valor que le indiquemos.

````python
# Obtener una clave que SÍ existe
poblacion_india = paisPoblacion.get('India', 'No disponible')
print("Población de India:", poblacion_india) 
````

````python
# Obtener una clave que NO existe
poblacion_brasil = paisPoblacion.get('Brasil', 'No disponible')
print("Población de Brasil:", poblacion_brasil)
````

5. **update(...)**: sirve para añadir nuevas parejas clave-valor o para actualizar las que ya existen.

````python
paisPoblacion.update({'Tanzania': 67438110, 'Pakistán': 230000000}) 
# Añade Tanzania (no existía) y actualiza la población de Pakistán (ya existía)
print("Diccionario después de update:", paisPoblacion)  
````

6. **pop(clave)**: elimina una pareja clave-valor del diccionario y nos devuelve el valor eliminado.

````python
poblacion_china = paisPoblacion.pop('China')
print("Diccionario después de pop:", paisPoblacion) 
print("Población de China eliminada:", poblacion_china)  # Salida: 1444216107
````

## Diccionarios anidados
Imagina que eres el responsable de informática de una empresa de muebles, y descubres que la información de los productos está guardada así:

````python
productos = ["Armarios", "Mesas", "Camas"]
stock = [100, 50, 75]
precios = [990, 590, 490]
````

¿Cuál crees que sería una mejor manera de almacenar estos datos? ¿Por qué?

### Solución
Una opción mucho mejor es usar un **diccionario anidado**, es decir, un diccionario dentro de otro diccionario:

````python
# Diccionario para almacenar datos de productos
muebles_info = {
    "Armarios": {"stock": 100, "precio": 990},
    "Mesas": {"stock": 50, "precio": 590},
    "Camas": {"stock": 75, "precio": 490}
}
````

Esto es mejor por estas razones:
* **Todo está junto**: si eliminas un producto, se elimina también su stock y su precio, sin dejar datos sueltos.
* **No hay líos con los índices**: en las listas, si te equivocas de posición puedes mezclar los datos de productos diferentes. Aquí eso no puede pasar.
* **No hay duplicados**: no puede haber dos productos con el mismo nombre.
* **Es más fácil de leer**: `muebles_info["Mesas"]["stock"]` se entiende mucho mejor que `productos[1]`, `stock[1]`.
