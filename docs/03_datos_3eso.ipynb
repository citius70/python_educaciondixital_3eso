# 📗 Tipos de datos en Python

## ¡Aprende a trabajar con información!

**Piensa en ello así:** Si tienes una caja para guardar zapatos, no puedes meter líquidos. Lo mismo pasa con los datos: cada tipo tiene sus propias reglas sobre qué puedes hacer con él.

Estos son los **tipos de datos básicos** que aprenderás hoy:

## Números enteros 🔢

Un **número entero** es un número completo, sin decimales. Puede ser positivo o negativo.

**Ejemplos del mundo real:**
- Tu edad: 15 años
- Puntos en un videojuego: 1250 puntos
- Temperatura bajo cero: -5 grados

En Python, los números enteros se llaman `int` (abreviatura de *integer*, que significa entero en inglés).

**Importante:** Los números se escriben **sin comillas**.

````python
# Ejemplos de números enteros
mi_edad = 15
puntos_juego = 1250
temperatura = -5

print(mi_edad)
print(puntos_juego)
````

:::{seealso} 💡 Truco: Incrementar valores
:class: dropdown

En programación es muy común **aumentar el valor de una variable**. Por ejemplo, cuando ganas puntos en un juego:

**Forma larga:**
```python
puntos = 10
puntos = puntos + 5  # Ahora puntos vale 15
```

**Forma corta (más usada):**
```python
puntos = 10
puntos += 5  # Significa: suma 5 a puntos. Ahora vale 15
```

También funciona con resta (`-=`), multiplicación (`*=`) y división (`/=`).
:::

````python
# Ejemplo: sistema de puntos en un videojuego
puntos = 0
print("Puntos iniciales:", puntos)

puntos += 50  # Completas un nivel
print("Después de completar nivel:", puntos)

puntos += 25  # Recoges una moneda especial
print("Después de recoger moneda:", puntos)

puntos -= 10  # Pierdes una vida
print("Después de perder vida:", puntos)
````

## Números decimales 🎯

Un **número decimal** (también llamado *flotante* o `float`) es un número que tiene parte decimal, separada por un **punto** (no coma).

**Ejemplos del mundo real:**
- Tu altura: 1.65 metros
- Precio de un producto: 19.99 euros
- Nota de un examen: 7.5

**¡Atención!** En Python usamos **punto** (`.`) como separador decimal, no coma.

````python
# Ejemplos de números decimales
mi_altura = 1.65
precio_auriculares = 29.99
nota_examen = 8.75

print("Mi altura es:", mi_altura, "metros")
print("Los auriculares cuestan:", precio_auriculares, "euros")
````

:::{warning} ⚠️ Cuidado con el separador decimal
En España escribimos 3,14 (con coma), pero en Python **siempre** se usa punto: `3.14`

❌ Incorrecto: `precio = 19,99`  
✅ Correcto: `precio = 19.99`
:::

## Textos 📝

Un **texto** (también llamado *cadena de texto* o `string`) es cualquier secuencia de letras, números o símbolos.

**Ejemplos del mundo real:**
- Tu nombre: "Ana"
- Un mensaje: "¡Hola, mundo!"
- Tu dirección de email: "ana@ejemplo.com"

**Regla importante:** Los textos **siempre** van entre comillas (simples `'` o dobles `"`).

````python
# Ejemplos de textos
mi_nombre = "Ana"
saludo = '¡Hola, mundo!'
cancion_favorita = "Blinding Lights"

print(saludo)
print("Mi canción favorita es:", cancion_favorita)
````

:::{tip} 💡 ¿Comillas simples o dobles?
Puedes usar cualquiera de las dos, funcionan igual:
- `nombre = "Ana"` ✅
- `nombre = 'Ana'` ✅

Lo importante es ser **consistente**: si empiezas con comillas dobles, termina con comillas dobles.
:::

## Verdadero o Falso (Booleanos) ✅❌

Un **booleano** (`bool`) es un tipo de dato que solo puede tener dos valores:
- `True` (Verdadero)
- `False` (Falso)

**Ejemplos del mundo real:**
- ¿Tienes mascota? → True o False
- ¿Has aprobado el examen? → True o False
- ¿Está lloviendo? → True o False

Los booleanos son muy útiles para que el programa tome decisiones (esto lo veremos más adelante).

````python
# Ejemplos de booleanos
tengo_mascota = True
ha_llovido_hoy = False
me_gusta_python = True

print("¿Tengo mascota?", tengo_mascota)
print("¿Me gusta Python?", me_gusta_python)
````

:::{warning} ⚠️ ¡Cuidado con las mayúsculas!
En Python, `True` y `False` **siempre** empiezan con mayúscula.

❌ Incorrecto: `aprobado = true`  
✅ Correcto: `aprobado = True`
:::

## 🔗 Unir textos (Concatenación)

A menudo necesitas **combinar varios textos** para crear mensajes completos. A esto se le llama **concatenación**.

En Python hay dos formas principales de hacerlo:

### Forma 1: Usar el símbolo `+`

El símbolo `+` **pega** los textos uno detrás de otro.

**Reglas importantes:**
- Solo puedes unir textos con textos
- Debes añadir los espacios manualmente
- Si quieres unir un número, primero debes convertirlo a texto con `str()`

````python
# Ejemplo 1: Unir textos
nombre = "Elena"
apellido = "García"

nombre_completo = nombre + " " + apellido  # Fíjate en el espacio " "
print(nombre_completo)

# Ejemplo 2: Unir textos y números
edad = 15
mensaje = "Tengo " + str(edad) + " años"  # Convertimos edad a texto
print(mensaje)
````

### Forma 2: Usar comas en `print()`

Dentro de `print()`, puedes separar varios elementos con comas (`,`).

**Ventajas:**
- Python añade los espacios automáticamente
- No necesitas convertir los números a texto
- Es más rápido para mostrar información

**Limitación:**
- Solo sirve para mostrar información, no para crear una variable nueva

````python
# Ejemplo con comas
nombre = "Elena"
edad = 15

# Python añade los espacios automáticamente
print("Me llamo", nombre, "y tengo", edad, "años")

# No necesitas str() para los números
puntos = 1250
print("Has conseguido", puntos, "puntos")
````

### ¿Cuándo usar cada forma?

| Método | ¿Crea variable? | ¿Necesita `str()`? | ¿Añade espacios? | ¿Cuándo usarlo? |
|--------|----------------|-------------------|------------------|------------------|
| **Operador `+`** | Sí | Sí | No, manual | Cuando quieres guardar el texto en una variable |
| **Coma `,`** | No | No | Sí, automático | Cuando solo quieres mostrar información |

````python
# Ejemplo comparativo
nombre = "Ana"
edad = 15

# Forma 1: Crear una variable con el mensaje
mensaje_guardado = "Me llamo " + nombre + " y tengo " + str(edad) + " años"
print(mensaje_guardado)

# Forma 2: Solo mostrar (más fácil)
print("Me llamo", nombre, "y tengo", edad, "años")
````

## 🔍 ¿Cómo saber el tipo de una variable?

A veces no estás seguro de qué tipo de dato contiene una variable. Para averiguarlo, usa la función `type()`.

````python
# Crear variables de diferentes tipos
mi_edad = 15
mi_nombre = "Ana"
mi_altura = 1.65
tengo_mascota = True

# Descubrir el tipo de cada variable
print("Tipo de mi_edad:", type(mi_edad))           # <class 'int'>
print("Tipo de mi_nombre:", type(mi_nombre))       # <class 'str'>
print("Tipo de mi_altura:", type(mi_altura))       # <class 'float'>
print("Tipo de tengo_mascota:", type(tengo_mascota)) # <class 'bool'>
````

:::{tip} 💡 Entendiendo la respuesta de `type()`
Cuando usas `type()`, Python te responde con algo como `<class 'int'>`. Esto significa:
- `int` → número entero
- `float` → número decimal
- `str` → texto (string)
- `bool` → verdadero/falso (boolean)

La palabra "class" significa "tipo" o "clase". No te preocupes por ella ahora.
:::

## 🔄 Transformar datos de un tipo a otro

A veces necesitas **cambiar el tipo de un dato**. Por ejemplo, convertir un número a texto para poder mostrarlo en un mensaje.

### Funciones de conversión

Python tiene funciones especiales para convertir tipos:

- `int()` → Convierte a número entero
- `float()` → Convierte a número decimal
- `str()` → Convierte a texto
- `bool()` → Convierte a verdadero/falso

### Ejemplo 1: De decimal a entero

````python
nota_decimal = 8.7
nota_entera = int(nota_decimal)

print("Nota original:", nota_decimal)    # 8.7
print("Nota como entero:", nota_entera)  # 8
````

:::{warning} ⚠️ Cuidado: Python no redondea
Cuando conviertes un decimal a entero, Python **elimina** la parte decimal (no la redondea).

- `int(8.7)` → `8` (no es 9)
- `int(3.99)` → `3` (no es 4)
:::

### Ejemplo 2: De número a texto

````python
puntos = 1250
puntos_texto = str(puntos)

# Ahora podemos usar el símbolo + para unir
mensaje = "Has conseguido " + puntos_texto + " puntos"
print(mensaje)
````

### Ejemplo 3: De texto a número

````python
edad_texto = "15"  # Esto es un texto, no un número
edad_numero = int(edad_texto)  # Convertimos a número

# Ahora podemos hacer operaciones matemáticas
edad_siguiente = edad_numero + 1
print("El año que viene tendrás", edad_siguiente, "años")
````

### ¿Qué conversiones funcionan?

**Conversiones que SÍ funcionan ✅:**
- `int("123")` → `123` (texto con números a entero)
- `float("3.14")` → `3.14` (texto con decimales a decimal)
- `str(42)` → `"42"` (número a texto)
- `int(3.9)` → `3` (decimal a entero)

**Conversiones que NO funcionan ❌:**
- `int("hola")` → ❌ ERROR (no puedes convertir letras a número)
- `float("abc")` → ❌ ERROR (no puedes convertir letras a decimal)
