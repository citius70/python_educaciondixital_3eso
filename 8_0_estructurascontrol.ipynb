# Introducción a las estructuras de control

Las **estructuras de control** en Python son estructuras que permiten dirigir la **ejecución del código según condiciones o repeticiones**. Esto incluye declaraciones: 
* Condicionales:
    * `if, else, elif`: ejecuta un código **si se cumple una condición**.
* Bucles:
    * `for`: repite un código un número **determinado de veces**.
    * `while`: repite un código un número **indeterminado** de veces mientras se cumple una condición.
